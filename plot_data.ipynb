```python
# %load plotdata.py
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from datetime import datetime
import pytz
init_notebook_mode(connected=True)
```

```python
tmp = np.loadtxt('data.log',usecols=1)
dates = []
with open('data.log','r') as f:
    for line in f:
        naive = datetime.fromisoformat(line.split()[0])
        tz_aware = naive.replace(tzinfo=pytz.utc)
        dates.append(tz_aware.astimezone(pytz.timezone('America/Chicago')))

data = [go.Scatter(x=dates, y=tmp)]
```

```python
iplot(data)
```
